```python
import numpy as np
import matplotlib.pyplot as plt

from mlhandmade.knn import KNNRegressor
from mlhandmade.linear_models import LinearRegressor

plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)
%matplotlib inline
```

```python
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
T = np.linspace(0, 5, 500)[:, np.newaxis]
y = np.sin(X).ravel()

y[::5] += 1 * (0.5 - np.random.rand(8))
```

```python
fig = plt.figure(figsize=(10,8))
k = 5
for i, weights in enumerate(["uniform", "distance", "gaussian"]):
    knn = KNNRegressor(algorithm="ball-tree", k=k, weights_mode=weights)
    knn.fit(X, y)
    y_ = knn.predict(T)

    plt.subplot(3, 1, i + 1)
    plt.scatter(X, y, color="darkorange", label="data")
    plt.plot(T, y_, color="navy", label="prediction")
    plt.axis("tight")
    plt.legend()
    plt.title("KNNRegressor (k = %i, weights = '%s')" % (k, weights))

plt.tight_layout()
plt.show()
```
