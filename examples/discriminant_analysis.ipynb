```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from mlhandmade.discriminant_analysis import (
    LinearDiscriminantAnalysis,
    QuadraticDiscriminantAnalysis
)
from mlhandmade.preprocessing import (
    standardize,
    binary,
    ordinal
)
from mlhandmade.plotting import(
    plot_decision_regions,
    scatter_plot,
)
```

```python
iris = pd.read_csv("../datasets/iris.csv")
iris.head()
```

```python
X = iris.iloc[:100, [0, 1]].values
y = iris.iloc[:100, -1].values

X = standardize(X)
y = binary(y)
fig = plt.figure(figsize=(10,8))
ax = scatter_plot(X, y, legend=2)
```

```python
fig = plt.figure(figsize=(12,12))
params = np.ones(3)
model = LinearDiscriminantAnalysis()
model.fit(X, y)
ax = plot_decision_regions(X=X, y=y, classifier=model, legend=2)
```

```python
fig = plt.figure(figsize=(12,12))
params = np.ones(3)
model = QuadraticDiscriminantAnalysis()
model.fit(X, y)
ax = plot_decision_regions(X=X, y=y, classifier=model, legend=2)
```
