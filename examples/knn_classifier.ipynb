```python
import pandas as pd
import matplotlib.pyplot as plt

from mlhandmade.knn import KNNClassifier
from mlhandmade.preprocessing import (
    binary,
    standardize,
    ordinal,
)
from mlhandmade.plotting import(
    scatter_plot,
    plot_decision_regions
)

plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)
%matplotlib inline
```

```python
iris = pd.read_csv("../datasets/iris.csv")
iris.head()
```

```python
X = iris.iloc[:, [0, 1]].values
y = iris.iloc[:, -1].values
```

```python
# X = standardize(X)
y = ordinal(y)
```

```python
%%time
fig = plt.figure(figsize=(10,8))
model = KNNClassifier(k=15, algorithm="kd-tree", weights_mode="distance")
model.fit(X, y)
ax = plot_decision_regions(X=X, y=y, classifier=model, legend=2)
ax.set_title("KNNClassifier with kd-tree")
plt.show()
```

```python
%%time
fig = plt.figure(figsize=(10,8))
model = KNNClassifier(k=15, algorithm="ball-tree", weights_mode="distance")
model.fit(X, y)
ax = plot_decision_regions(X=X, y=y, classifier=model, legend=2)
ax.set_title("KNNClassifier with ball-tree")
plt.show()
```

```python
%%time
fig = plt.figure(figsize=(10,8))
model = KNNClassifier(k=15, algorithm="brute", weights_mode="distance")
model.fit(X, y)
ax = plot_decision_regions(X=X, y=y, classifier=model, legend=2)
ax.set_title("KNNClassifier with brute-force")
plt.show()
```
