```python
from mlhandmade.clustering import KMeans
from mlhandmade.preprocessing import ordinal
from mlhandmade.plotting import scatter_plot
from mlhandmade.model_selection import accuracy_score

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(random_state=0)
# X = standardize(X)
model = KMeans(n_clusters=3, init="k-means++", max_iter=300, tol=1e-4)
model.fit(X)
kmeans_accuracy = accuracy_score(y, model.labels_, normalize=True)

fig, axs = plt.subplots(1, 2, figsize=(16,8))
scatter_plot(X, y, ax=axs[0], legend = 2)
scatter_plot(X, model.labels_, ax=axs[1], legend = 2)
axs[0].set_title("True labels")
axs[1].set_title(f"KMeans labels, accuracy: {kmeans_accuracy}")
plt.show()
```
